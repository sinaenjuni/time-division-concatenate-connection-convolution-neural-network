````python

def spatial_pyramid_pool(x, k, bn_axis, name):
    size = int(backend.int_shape(x)[-2])

    pools = []
    for i in range(k):
        level = 2 ** i
        kernal_size = math.ceil(size / level)
        stride = math.floor(size/level)

        pool = layers.MaxPooling1D(kernal_size, strides=stride, name=name + '_spatial_pyramid_pool')(x)
        pools.append(pool)

    spp  = layers.Concatenate(axis=bn_axis, name=name + '_spp_concat')(pools)

    return spp
````
